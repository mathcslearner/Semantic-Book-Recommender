```python
# Import Data

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
books.describe(include = "all")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot missing values heatmap

ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np 

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2026 - books["published_year"]
```

```python
# Investigate correlation of missing fields

columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style = "white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
# Remove books with missing fields

book_no_missing = books[~(books["description"].isna()) & ~(books["num_pages"].isna()) & ~(books["average_rating"].isna()) & ~(books["published_year"].isna())]
book_no_missing
```

```python
book_no_missing.describe(include="all")
```

```python
book_no_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_no_missing["words_in_description"] = book_no_missing["description"].str.split().str.len()
```

```python
# Analyze word count in descriptions

sns.histplot(book_no_missing["words_in_description"], bins=500, kde=False)
plt.title("Distribution of Word Count in Descriptions")
plt.xlabel("Number of Words")
plt.ylabel("Count")
plt.show()
```

```python
# Remove books with description shorter than 25 words

book_no_missing_25_words = book_no_missing[book_no_missing["words_in_description"] >= 25]
book_no_missing_25_words
```

```python
# Join title and subtitle together

book_no_missing_25_words["title_and_subtitle"] = (np.where(book_no_missing_25_words["subtitle"].isna(), 
                                                  book_no_missing_25_words["title"], 
                                                  book_no_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)))
```

```python
book_no_missing_25_words
```

```python
# Join ISBN and description together

book_no_missing_25_words["tagged_description"] = book_no_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
book_no_missing_25_words
```

```python
# Drop the columns no longer needed

(book_no_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned.csv", index = False))
```
